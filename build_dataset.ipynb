```python
import numpy as np
import pandas as pd
```

```python
newest = 2018
```

```python
ba18 = pd.read_csv('./data/precio-de-terrenos-2018.csv')
ba17 = pd.read_csv('./data/precio-de-terrenos-2017.csv', delimiter=';')
ba16 = pd.read_csv('./data/precio-de-terrenos-2016.csv', delimiter=';')
ba15 = pd.read_csv('./data/precio-de-terrenos-2015.csv', delimiter=';')
ba14 = pd.read_csv('./data/precio-de-terrenos-2014.csv', delimiter=';')

dfs = [ba18, ba17, ba16, ba15, ba14]
```

```python
names = {'lat': 'LAT',
        'long': 'LON',
        'm2total': 'M2',
        'preciousd': 'USD',
        'barrio': 'NEIGHBORHOOD',
        'comuna': 'COMMUNE',
        'X': 'LAT',
        'Y': 'LON', 
        'DOLARES': 'USD', 
        'BARRIO': 'NEIGHBORHOOD',
        'BARRIOS': 'NEIGHBORHOOD',
        'COMUNA': 'COMMUNE'}
```

```python
for i, df in enumerate(dfs):
    df.rename(columns=names, inplace=True)
    drop_cols = set(df.columns) - set([v for v in names.values()])
    df.drop(columns=list(drop_cols), inplace=True)
    df['YEAR'] = newest - i
    
df = pd.concat(dfs)
```

```python
fix_neighborhoods = {'MONTSERRAT': 'MONSERRAT', 
                     'NUÃ‘EZ': 'NUÑEZ', 
                     'NU?ÆEZ': 'NUÑEZ'}

df['NEIGHBORHOOD'].replace(fix_neighborhoods, inplace=True)
```

```python
fix_communes = {'COMUNA 1': 1,
                'COMUNA 2': 1,
                'COMUNA 3': 3,
                'COMUNA 4': 4,
                'COMUNA 5': 5,
                'COMUNA 6': 6,
                'COMUNA 7': 7,
                'COMUNA 8': 8,
                'COMUNA 9': 9,
                'COMUNA 10': 10,
                'COMUNA 11': 11,
                'COMUNA 12': 12,
                'COMUNA 13': 13,
                'COMUNA 14': 14,
                'COMUNA 15': 15}

df['COMMUNE'].replace(fix_communes, inplace=True)
```

```python
df.dropna(inplace=True)
df.drop_duplicates(subset=['YEAR', 'NEIGHBORHOOD', 'M2', 'USD'], inplace=True)
df.sort_values(by=['YEAR', 'COMMUNE', 'NEIGHBORHOOD', 'USD'], inplace=True)
```

```python
df['LAT'].replace({' ': np.nan}, inplace=True)
df['LON'].replace({' ': np.nan}, inplace=True)
df['LAT'] = df['LAT'].astype(np.float)
df['LON'] = df['LON'].astype(np.float)
df['COMMUNE'] = df['COMMUNE'].astype(int)
```
